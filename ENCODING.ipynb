```python
data = {
    "college":["Adani","Nirma", "LPU", "UPES", "PDPU","Nirma","LPU","Adani","PDPU","DAIICT"],
    "target":[270, 350, 400, 300,501, 360, 410, 280, 510, 330]
}
```

```python
df = pd.DataFrame(data)
display(df)
```

## Label Encoding


```python
label_encoder = LabelEncoder()
df['college_label_encoded'] = label_encoder.fit_transform(df['college'])
print("college: labels as 0: Adani, 1: Nirma, 2: LPU, 3: UPES, 4: PDPU, 5: DAIICT")
display("label encoded \n", df)
```

## One-Hot Encoding

```python
# One-Hot Encoding using pandas get_dummies
one_hot = pd.get_dummies(df, columns=['college'], prefix='college')
df_one_hot = pd.concat([df, one_hot], axis=1)
display("one hot encoded:\n", df_one_hot)
```

## Binary Encoding

```python
binary_encoder = BinaryEncoder(cols=['college'])
df_binary = binary_encoder.fit_transform(df["college"])
df_binary_encoded = pd.concat([df, df_binary], axis=1)
display("\nbinary encoded:\n", df_binary_encoded)
```

## Mean Encoding

```python
mean_encoded = df.groupby('college')['target'].mean()
df['college_mean_encoded'] = df['college'].map(mean_encoded)
display("\nmean encoded:\n", df)
```

## Frequency Encoding

```python
freq_encoded = df['college'].value_counts() / len(df)
df['college_freq_encoded'] = df['college'].map(freq_encoded)
display("\nfrequency encoded:\n", df)
```
