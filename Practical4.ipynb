```python
for col in ['age','fare','parch']: 
            print(f"\n columns: {col}")
            print(f"\n mean: {df[col].mean}")
            print(f"\n median: {df[col].median}")
            print(f"\n Standard deviation: {df[col].std}")
            print(f"\n Q1: {df[col].quantile(0.25)}")
            print(f"\n Q2: {df[col].quantile(0.50)}")
            print(f"\n Q3: {df[col].quantile(0.75)}")
```

```python

import matplotlib.pyplot as plt
plt.hist(df["age"])
plt.title("AGE")
plt.xlabel("age")
plt.ylabel("freq")
plt.show()
```

```python
plt.hist(df["fare"], color ="red")
plt.title("fare")
plt.xlabel("fare")
plt.ylabel("freq")
plt.show()
```
